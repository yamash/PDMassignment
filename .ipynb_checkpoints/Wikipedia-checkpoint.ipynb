### V- card info

```python
def feature_names(ht):
    feature_name = []
    if ht is not None:
        for h in ht.find_all('th'):
            feature_name.append(h.get_text())
    return feature_name
```

```python
def feature_values(html):
    values = []
    if html is not None:
        for a in html.find_all('td'):
            if a.get_text() != '\n':
                values.append(a.get_text().strip())
    return values
```

```python
def website(html):
    if html is not None:
        try:
            return html.find_all('a')[-1].get('href')
        except:
            pass
    else:
        return None
```

```python
def infobox(html):
    info = []
    values = []
    dictionary = []
    dic = {}
    wiki = pd.DataFrame()
    if html is not None:
        for tr in html.find_all('tr'):
            if len(list(tr.children)) == 2:
                try:
                    dic.update({tr.find('th').get_text(): tr.find('td').get_text().strip()})
                except:
                    dic.update({'None': None})
                    pass
            else:
                dic.update({'None': None})
    else:
        dic.update({'None': None})
                    
    return dic
```

```python
def saveJSONLine(array, filename):
    print('saving ' + filename)
    with open(filename, 'w') as jsonfile:
        count = 0
        for r in array:
            line = json.dumps(r)
            jsonfile.write(line + '\n')
            if count % 10000 == 0:
                print('.', end="")
            count += 1
    print('done!')
```
